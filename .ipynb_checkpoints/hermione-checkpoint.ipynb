## TFIDF Exploration

```python
vocab_book = hpe.extract_vocab(corpus)[1] # new vocab table by book
vocab_book = vocab_book.drop(['-bodied', '-couldn', '-joke-about-things-like-', '-know-', '-two', '-year', 'yellow'], level=2)
```

```python
def comp_tfidf(bow):
    
    dtcm = bow.n.unstack().fillna(0).astype(int)
    tf = dtcm.T/dtcm.T.sum()
    tf = tf.T
    df = dtcm.astype(bool).sum()
    n = dtcm.shape[0]
    idf = np.log2(n/df)
    tfidf = tf*idf
    
    return tfidf.stack(), tfidf
```

```python
tfidfs = comp_tfidf(vocab_book)
tfidf_list = tfidfs[0]
tfidf = tfidfs[1]
```

```python
vocab_book['tfidf'] = tfidf_list
```

```python
vocab_book.tail()
```

```python
hermione_qualities = ['hermione', 'granger', 'smart', 'intelligent', 'clever', 'logical', 'bossy'] #https://en.wikipedia.org/wiki/Hermione_Granger#:~:text=Hermione's%20most%20prominent%20features%20include,%2C%20the%20%22Dark%20Lord%22.
hermione_vocab = vocab_book.loc[(slice(None), slice(None), hermione_qualities), :].reset_index(level=2, drop=True)
```

```python
hermione_mean_tfidf = pd.DataFrame(hermione_vocab.groupby(['book_id', 'chap_num']).tfidf.mean()) # total hermione language tfidf
```

```python
# plot hermione mean tfidf
plot_herm = hermione_mean_tfidf.reset_index().melt(id_vars=['chap_num', 'tfidf'])
plot_herm.columns = ['chap_num', 'tfidf', 'var', 'book']
```

```python
herm_by_book = sns.relplot(data=plot_herm, kind='line', col='book', col_wrap=4, x='chap_num', y='tfidf',
            height=3.75, facet_kws={'sharey': False, 'sharex': True})
herm_by_book.fig.subplots_adjust(top=.9)
herm_by_book.fig.suptitle('TFIDF of Hermione Qualities Across Books')
```

The above plot shows distribution of the mean TFIDF of Hermione associated language across chapters, separated by book.

TFIDF serves as measure of importance or relevance. In the plot above, we see that Hermione (and words commonly associated with Hermione) are typically the most relevant in the middle of each of Rowling's books because the TFIDF at these points peaks. Because the middle of a novel usually denotes the climax, we surmise that Hermione is essential to climactic events throughout all books in the Harry Potter series.

```python
# hermione importance over time by book - see the peaks
fig = px.line(plot_herm, x='chap_num', y='tfidf', color='book', markers=True, title='Mean TFIDF of Hermione Qualities over Time')
fig.show()
```

The peaks indicate moments where hermione is most important/relevant according to TFIDF. We see that there are TFIDF peaks in books 3, 4, 5, 6, and 7. Hermione seems less important in books 1 and 2. This makes sense because in the beginning of the series, Harry and Ron are unsure about Hermione and whether they like her. At the end of book 5, Hermione peaks in importance [SAY WHY]. ## Hierarchical

```python
PAIRS = pd.DataFrame(index=pd.MultiIndex.from_product([lib.index.tolist(), lib.index.tolist()])).reset_index()
PAIRS = PAIRS[PAIRS.level_0 < PAIRS.level_1].set_index(['level_0','level_1'])
PAIRS.index.names = ['doc_a', 'doc_b']
```

```python
imp_book = tfidf_chap.groupby(['book_id', 'term_str']).sum()
imp_book = imp_book.unstack(fill_value=0)
```

```python
imp_book
```

```python
L0 = imp_book.astype('bool').astype('int') # Binary (Pseudo L)
L1 = imp_book.apply(lambda x: x / x.sum(), 1) # Probabilistic
L2 = imp_book.apply(lambda x: x / norm(x), 1) # Pythagorean / Euclidean
```

```python
PAIRS['cityblock'] = pdist(imp_book, 'cityblock')
PAIRS['euclidean'] = pdist(imp_book, 'euclidean')
PAIRS['cosine'] = pdist(imp_book, 'cosine')
PAIRS['jaccard'] = pdist(L0, 'jaccard')
PAIRS['dice'] = pdist(L0, 'dice')
PAIRS['js'] = pdist(L1, 'jensenshannon') 
```

```python
PAIRS
```

```python
colors = "YlGnBu"
# PAIRS.style.background_gradient(colors)
```

```python
PAIRS['cosine'].to_frame().unstack(fill_value=0).style.background_gradient(colors)
```

```python
PAIRS['cosine'].to_frame().sort_values(by='cosine', ascending=False).head(3)
```

The three greatest correlations according to cosine are found between books 2 and 7, books 1 and 6, and books 2 and 3. It is interesting that books where Hermione is seen a significant character (books 3, 6, and 7) are correlated according to cosine with books where Hermione is not as relevant (books 1 and 2).

## Hierarchical

```python
import scipy.cluster.hierarchy as sch

def hca(sims, LIB, linkage_method='complete', color_thresh=.3, figsize=(10, 10)):
    tree = sch.linkage(sims, method=linkage_method)
    labels = LIB.title.values
    plt.figure()
    fig, axes = plt.subplots(figsize=figsize)
    dendrogram = sch.dendrogram(tree, 
                                labels=labels, 
                                orientation="left", 
                                count_sort=True,
                                distance_sort=True,
                                above_threshold_color='.75',
                                color_threshold=color_thresh
                               )
    plt.tick_params(axis='both', which='major', labelsize=14)
```

```python
hca(PAIRS.cosine, lib, linkage_method='ward', color_thresh=2)
```

The hierarchical cluster above shows which books are most similar to each other in the Harry Potter series according to cosine similarity and ward linkage methods. We see that books 7 and book 5 are the most similarly related. There is an overarching divide between books 1, 2, and 3 and books 4, 5, 6, and 7. This suggests that after book 3, Rowling enacts a semantic divide in the series.

## Topic Modeling 

```python
ngram_range = [1,2]
n_terms = 4000
n_topics = 10
max_iter = 20
n_top_terms = 15
```

```python
phi = pd.read_csv('output/LDA-PHI.csv')

phi.index.name = 'topic_id'
phi.columns.name  = 'term_str'
```

```python
phi
```

```python
DOCS = corpus[corpus.pos.str.match(r'^NNS?$')]\
    .groupby(['book_id', 'chap_num']).term_str\
    .apply(lambda x: ' '.join(x))\
    .to_frame()\
    .rename(columns={'term_str':'doc_str'})
```

```python
DOCS
```

```python
count_engine = CountVectorizer(max_features=n_terms, ngram_range=ngram_range, stop_words='english')
count_model = count_engine.fit_transform(DOCS.doc_str)
TERMS = count_engine.get_feature_names_out()
```

```python
DTM = pd.DataFrame(count_model.toarray(), index=DOCS.index, columns=TERMS)
```

```python
vocab['doc_count'] = DTM.astype('bool').astype('int').sum()
DOCS['term_count'] = DTM.sum(1)
```

```python
lda_engine = LDA(n_components=n_topics, max_iter=max_iter, learning_offset=50., random_state=0) # gen model
```

```python
TNAMES = [f"T{str(x).zfill(len(str(n_topics)))}" for x in range(n_topics)] # name topics numerically
```

```python
lda_model = lda_engine.fit_transform(count_model)
```

```python
THETA = pd.DataFrame(lda_model, index=DOCS.index)
THETA.columns.name = 'topic_id'
THETA.columns = TNAMES
```

```python
THETA.sample(20).style.background_gradient(cmap=colors, axis=None)
```

```python
PHI = pd.DataFrame(lda_engine.components_, columns=TERMS, index=TNAMES)
PHI.index.name = 'topic_id'
PHI.columns.name  = 'term_str'
```

```python
PHI.T.sample(20).style.background_gradient(cmap=colors, axis=None)
```

```python
TOPICS = PHI.stack().to_frame('topic_weight').groupby('topic_id')\
    .apply(lambda x: x.sort_values('topic_weight', ascending=False)\
        .head(n_top_terms).reset_index().drop('topic_id', axis=1)['term_str'])
```

```python
TOPICS
```

```python
TOPICS['label'] = TOPICS.apply(lambda x: x.name + ' ' + ', '.join(x[:n_top_terms]), 1)
```

```python
TOPICS.label.values
```

```python
TOPICS['doc_weight_sum'] = THETA.sum()
TOPICS['term_freq'] = PHI.sum(1) / PHI.sum(1).sum()
```

```python
TOPICS
```

```python
doc_freq = TOPICS.sort_values('doc_weight_sum', ascending=True).plot.barh(y='doc_weight_sum', x='label', figsize=(5, n_topics/2));
```

Hermione is shown in the the second most important topic in terms of document weight. This means that throughout all books, out of all topics, school (or Hogwarts) is the second most common topic, and it happens to contain Hermione as the only character in the top 15 words associated with the topic. The topic of schooling suggests that a key value throughout the Harry Potter series is intelligence, and it is a compliment that Hermione is portrayed as the most intelligent character.

Additionally, T03 contains words pertaining to women/girls and defense against the dark arts. Although this first seems promising, this is likely a topic regarding a teaching moment when Harry was instructing a few girls on defense against the dark arts, and Harry is interested in one of his pupils. That being said, women are relevant to the topic, but not in the context of feminine strength.
